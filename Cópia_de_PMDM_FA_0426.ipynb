```python
import pandas as pd
import time
import numpy as np
from sklearn.model_selection import RepeatedKFold
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.feature_selection import SequentialFeatureSelector, RFE
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LassoCV
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.feature_selection import VarianceThreshold
from sklearn.impute import KNNImputer


def cv_score(models, X_encoded, y):
    # Creating a list to store all score results:
    rkf = RepeatedKFold(n_splits=5, n_repeats=2, random_state=42)
    acc_score_train = []
    acc_score_val = []
    precision_score_train = []
    precision_score_val = []
    recall_score_train = []
    recall_score_val = []  
    f1_score_train = []
    f1_score_val = []
    timer = []  # timer
    model_names = []

    for model_info in models:  
        model, model_name = model_info 
        # Get indexes
        for train_index, val_index in rkf.split(X_encoded, y):
            X_train, X_val = X_encoded.iloc[train_index], X_encoded.iloc[val_index]
            y_train, y_val = y.iloc[train_index], y.iloc[val_index]

            # Counting time
            begin = time.perf_counter()


            # Preprocessing

            #Missing Values

          # dealing with nan in non_metric_features

      
            modes_1 = X_train[non_metric_features].mode().loc[0]

            X_train.fillna(modes_1, inplace=True)

            X_val.fillna(modes_1, inplace=True)

            #Dealing with nan in boolean variables
                      
            modes_b = X_train[df_booleans].mode().loc[0]
                        
            X_train.fillna(modes_b, inplace=True)
                                    
            X_val.fillna(modes_b, inplace=True)

#Encoding the categorical variables
            X_train_encoded = pd.get_dummies(X_train, columns=['Competition', 'Region','Education','Sex', 'Income','Age group'])

            X_val_encoded = pd.get_dummies(X_val, columns=['Competition', 'Region','Education','Sex', 'Income','Age group'])            




            #KNN 
            
            imputer = KNNImputer(n_neighbors=5, weights="uniform")
            #X_train[df_training_type] = imputer.fit_transform(X_train[df_training_type])
            #X_val[df_training_type] = imputer.fit_transform(X_val[df_training_type])


            X_train_imputed = pd.DataFrame(imputer.fit_transform(X_train_encoded), columns=X_train_encoded.columns, index=X_train_encoded.index)
            X_val_imputed = pd.DataFrame(imputer.transform(X_val_encoded), columns=X_val_encoded.columns, index=X_val_encoded.index)

            X_train_encoded[df_training_type] = X_train_imputed[df_training_type]
            #X_train = X_train_imputed
            X_val_encoded[df_training_type] = X_val_imputed[df_training_type]
            #X_val = X_val_imputed





            # Handling outliers

            filters = (
                (X_train_encoded['Strength training'] <= 9500) &
                (X_train_encoded['Sand training'] <= 2980) &
                (X_train_encoded['Supplements'] <= 2800) &
                (X_train_encoded['Squad training'] <= 220) &
                (X_train_encoded['Other training'] <= 300) &
                (X_train_encoded['Plyometric training'] <= 80) &
                (X_train_encoded['Sport-specific training'] <= 1000)
            )

            X_train = X_train_encoded[filters]
            y_train = y_train[filters]

      


            # Scaling
            feature_scaler = RobustScaler().fit(X_train)
            X_train_scaled = pd.DataFrame(feature_scaler.transform(X_train), columns=X_train.columns, index=X_train.index)
            X_val_scaled = pd.DataFrame(feature_scaler.transform(X_val_encoded), columns=X_val_encoded.columns, index=X_val_encoded.index)



            # Remove constant features
            constant_filter = VarianceThreshold()
            constant_filter.fit(X_train_scaled)
            X_train_scaled = X_train_scaled.loc[:, constant_filter.get_support()]
            X_val_scaled = X_val_scaled.loc[:, constant_filter.get_support()]


            # Feature Selection using RFE with cross-validation

            rfe = RFE(estimator=LogisticRegression(), n_features_to_select=10)
            rfe.fit(X_train_scaled, y_train)
            X_train_selected = rfe.transform(X_train_scaled)
            X_val_selected = rfe.transform(X_val_scaled)




          #tHIS IS A DIFFERENT METHOD FOR FEATURE SELECTION

           # if isinstance(model, LassoCV):
                # Lasso feature selection
           #     lasso = LassoCV(cv=4, random_state=42)
           #    lasso.fit(X_train_scaled, y_train)
           #     model_selector = SelectFromModel(lasso, prefit=True)
          #      X_train_selected = model_selector.transform(X_train_scaled)
          #      X_val_selected = model_selector.transform(X_val_scaled)
          #  else:
                # SelectKBest feature selection
          #      selector = SelectKBest(score_func=f_classif, k=10)
          #      X_train_selected = selector.fit_transform(X_train_scaled, y_train)
          #      X_val_selected = selector.transform(X_val_scaled)



            # Fit model to the data
            model.fit(X_train_selected, y_train)

            # Finish counting time
            end = time.perf_counter()

            # Predict for the train
            pred_train = model.predict(X_train_selected)

            # Predict for val
            pred_val = model.predict(X_val_selected)

            # Create variables for scores
            acc_train = accuracy_score(y_train, pred_train)
            acc_val = accuracy_score(y_val, pred_val)
            precision_train = precision_score(y_train, pred_train)
            precision_val = precision_score(y_val, pred_val)
            recall_train = recall_score(y_train, pred_train)
            recall_val = recall_score(y_val, pred_val)
            f1_train = f1_score(y_train, pred_train)
            f1_val = f1_score(y_val, pred_val)

            # Storing the scores in the corresponding different lists
            acc_score_train.append(acc_train)
            acc_score_val.append(acc_val)
            precision_score_train.append(precision_train)
            precision_score_val.append(precision_val)
            recall_score_train.append(recall_train)
            recall_score_val.append(recall_val)
            f1_score_train.append(f1_train)
            f1_score_val.append(f1_val)
            timer.append(end - begin)
            model_names.append(model_name)

    # Store results in DataFrame

    model_results = pd.DataFrame(data={
        'Model': model_names, 
        'Train_Acc': acc_score_train,
        'Val_acc': acc_score_val,
        'Train_Precision': precision_score_train,
        'Val_Precision': precision_score_val,
        'Train_Recall': recall_score_train,
        'Val_Recall': recall_score_val,
        'Train_F1': f1_score_train,
        'Val_F1': f1_score_val,
        'Time': timer
    }, index=[f'Fold {i}' for i in range(1, len(model_names) + 1)])

  
    return model_results


# Apply the models

models = [
    (DecisionTreeClassifier(max_depth=12), 'Decision Tree'),  
    (KNeighborsClassifier(), 'KNN'),
    (RandomForestClassifier(max_depth=12), 'Random Forest'),
    (GaussianNB(), 'Naive Bayes'),
    (AdaBoostClassifier(), 'Boosting'),
    (BaggingClassifier(), "Bagging")
]

results = cv_score(models, X_encoded, y)
print(results)
```

The timer in the code is used to measure the time taken for each iteration of the cross-validation loop. It captures the execution time from the start of feature scaling, feature selection, model training, and prediction to the end of those operations.

By measuring the time, you can get an idea of the computational efficiency of your code and compare the execution times of different models or different configurations. It can be useful for performance analysis and identifying potential bottlenecks in your code. """"To select among metric features, we will use Recursive Feature Elimination (RFE), which is Wrapper a technique that selects the most important features to keep. Logistic Regression will be the base estimator.

Remember, from previous practical class, that the sklearn implementation of RFE requires us to tell RFE the number of features we want to keep. Usually, that is knowledge that you do not have (or at least shouldn't) apriori.

You can, however, create a loop to check the score of the estimator using a different possible numbers of features.
Below you have an example that will check the score of the estimator by chosing between 2 and 15 variables. Another option would be to use RFECV instead."""

```python
X_train[df_booleans + metric]
```

```python
from sklearn.metrics import f1_score
```

```python
#import logistic regression and RFE, and create a classifier called model
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE

model_RFE = LogisticRegression()
```

```python
#Run the following loop to determine the optimal number of numeric features to select according to a RFE on the base estimator Logistic Regression on X_train_scaled.
# Number of features to keep
nof_list_RFE =np.arange(2,20)            
high_score_RFE =0
# Variable to store the optimum features
nof_RFE =0           

#list of scores
score_list_RFE =[]

for n in range(len(nof_list_RFE)):
    model_RFE = LogisticRegression()
    
    #create instance of RFE
    rfe = RFE(model_RFE,n_features_to_select = nof_list_RFE[n])
    
    #fit RFE to training data
    X_train_rfe = rfe.fit_transform(X_train_scaled,y_train)
    
    #Apply the same RFE transformation to the test data
    X_test_rfe = rfe.transform(X_test_scaled)
    
    #Create and train instance of Logistic Regression
    model_RFE.fit(X_train_rfe,y_train)
    
    #predict 
    y_pred = model_RFE.predict(X_test_rfe)
    
    #store f1score in variable
    f_score_RFE = f1_score(y_test, y_pred)
    
    score_list_RFE.append(f_score_RFE)
    
    #compare against previous best performance
    if(f_score_RFE > high_score_RFE):
        high_score_RFE = f_score_RFE
        nof_RFE = nof_list_RFE[n]
print("Optimum number of features: %d" %nof_RFE)
print("Score with %d features: %f" % (nof_RFE, high_score_RFE))
```

```python
#Initialize RFE
#Parameters: Estimator = model
#            features selected = 10

rfe = RFE(estimator = model_RFE, n_features_to_select = 16)
```

 SequentialFeatureSelector (SFS). SFS is a greedy procedure where, at each iteration, we choose the best new feature to add to our selected features based a cross-validation score. That is, we start with 0 features and choose the best single feature with the highest score. The procedure is repeated until we reach the desired number of selected features.
